# Exercise 6

Two datasets related to the acquisition and foundation of *startups* at a global level have been obtained, the general objective is to carry out an exploratory data analysis on the common elements of both datasets. These are the objectives to be achieved:
* Convert the **funds.csv** and **funding_rounds.csv** files to pandas dataframes.
* Search for a common element of both dataframes and convert them into one.
* Remove 'id', 'source_url', 'source_description', 'updated_at' columns.
* Shows the number of nulls that each column of the dataframe has and what percentage the number of nulls represents within each column.
* Changes all null 'raised_currency_dode' values to 'USD'.
* Removes all other nulls from the dataset.
* Shows the percentage of different values of 'raised_currency_dode'.
* Create a subset without the USD value and display a bar chart.
* Create a new subset by filtering from the dataset all rows below 10M and display the histogram.
* With the set of the previous section. Check out the contains() function https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.str.contains.html filters the dataset by startup creation dates in 2000

# SOLUTION

Convert **funds.csv** and **funding_rounds.csv** files to pandas dataframes

```python
import pandas as pd
```

```python
# Read both csv
funds = pd.read_csv('funds.csv')
f_rounds = pd.read_csv('funding_rounds.csv')
```

```python
# Header
funds.head(3)
```

```python
# Header
f_rounds.head(3)
```

```python
print("Funds columns ", funds.columns)
print("Funding rounds columns ", f_rounds.columns)
```

```python
print("Funds shape ", funds.shape)
print("Funding rounds shape ", f_rounds.shape)
```

Find a common element of both dataframes and convert them into one.

```python
df = pd.concat([f_rounds, funds], ignore_index=True, join='inner')
```

```python
df.head(3)
```

```python
df.shape
```

Remove 'id', 'source_url', 'source_description', 'updated_at' columns.

```python
items_to_drop = ['id', 'source_url', 'source_description', 'updated_at']
```

```python
df = df.drop(items_to_drop, axis=1)
```

Shows the number of nulls that each column of the dataframe has and what percentage the number of nulls represents within each column.

```python
def show_nulls(df):
    # Total nulls
    total_nan = df.isnull().sum()
    # Amount of nulls
    perc_nan = total_nan/len(df)*100
    
    return total_nan, perc_nan
```

```python
print('Total nulls ', show_nulls(df=df)[0])
```

```python
print('Nulls percentage ', show_nulls(df=df)[1])
```

Changes all null 'raised_currency_dode' values ​​to 'USD'.

```python
df['raised_currency_code'] = df['raised_currency_code'].fillna('USD')
```

Remove the resto of nulls

```python
df = df.dropna()
```

Shows the percentage of the different values of 'raised_currency_dode'

```python
# Primero atendemos a los valores únicos del tipo de moneda.
df['raised_currency_code'].unique()
```

```python
print((pd.crosstab(index=df["raised_currency_code"], columns="count"))/len(df) * 100)
```

Create a subset without the USD value and display a bar chart.

```python
no_usd = df[df['raised_currency_code'] != 'USD'].copy()
```

```python
groups = no_usd.groupby(["raised_currency_code"]).count()['object_id']
```

```python
groups.values
```

```python
n_groups = len(groups)

index = np.arange(n_groups)

max_ = plt.bar(index, groups.values, color='tomato', 
                 label='Currency')

plt.xlabel('Currency Type')
plt.ylabel('Amount')
plt.title('Amount by currency by startups')
plt.xticks(index , groups.keys())
plt.legend()

plt.show()
```

Create a new subset by filtering out all rows below 10M from the dataset and display the histogram.

```python
less_tenm = df[df['raised_amount'] <= 10000000].copy()
less_tenm.describe()
```

```python
less_tenm.plot(kind = "hist")
```
